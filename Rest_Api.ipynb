Problem 1 — Basic GET request Make a GET request to https://httpbin.org/get and print the status code and the url field from the JSON response.

Status: 200
URL: https://httpbin.org/get

```python
import requests
response = requests.get('https://httpbin.org/get')
print("Status:", response.status_code)
print("URL:", response.url)
```

Problem 2 — Read JSON response Make a GET request to https://httpbin.org/json and print the title field from the response JSON.

Wake up to WonderWidgets!

```python
import json

response_json = requests.get('https://httpbin.org/json')
data=response_json.json()
print(json.dumps(data, indent=4))
print(data['slideshow']['title'])
print(response_json.json()['slideshow']['slides'][0]['title'])
```

Problem 3 — Query parameters Make a GET request to https://httpbin.org/get passing name=Arun and city=Chennai as query parameters. Print the args from the response.

{'city': 'Chennai', 'name': 'Arun'}

```python
parameter={"name":"Arun",
           "city":"Chennai"}
response_json = requests.get('https://httpbin.org/get',params=parameter)
data=response_json.json()
print(response_json.url)
print(json.dumps(data, indent=4))
print(data["args"])
```

Problem 4 — POST with JSON body Send a POST request to https://httpbin.org/post with a JSON body containing title, severity, and status. Print the json field from the response.

{'severity': 'high', 'status': 'open', 'title': 'Login broken'}

```python
payload={'severity': 'high', 'status': 'open', 'title': 'Login broken'}
response=requests.post('https://httpbin.org/post',json=payload) # https://httpbin.org/post echoes back what you sent
#Use the json= parameter in requests.post() to send a JSON body.
data=response.json()
print(data['json']) # contains your original JSON body
```
